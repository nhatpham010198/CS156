```python
from PIL import Image
import PIL.ImageOps
from resizeimage import resizeimage
from collections import defaultdict
from glob import glob
from random import shuffle, seed
import numpy as np
import pylab as pl
import pandas as pd
import re
from sklearn.decomposition import PCA as RandomizedPCA
from sklearn.linear_model import LogisticRegression
```

```python
def resize(path_to_image, width, height):
        """Resize the image and save it again.
        :param path_to_image: os.path
        :param width: int
        :param height: int
        :return: None
        """
        try:
            fd_img = open(path_to_image, 'rb')
            img = Image.open(fd_img)
            img = resizeimage.resize_cover(img, [int(width), int(height)])
            img.save(path_to_image, img.format)
            fd_img.close() 
        except:
            pass
```

```python
import cv2
def flatten(file):
    image = PIL.Image.open(i)
    width, height = image.size
    string_img = cv2.imread(file).reshape(1,width*height*3)
    return(string_img[0])
```

```python
men_files = glob("/Users/nhatpham/Downloads/Man's Clothing/*.JPEG")
women_files = glob("/Users/nhatpham/Downloads/Man's Clothing/*.JPEG")
raw_data = [[flatten(filename),'men',filename] for filename in men_files] + \
           [[flatten(filename),'women',filename] for filename in women_files]

# randomly order the data
seed(0)
shuffle(raw_data)
```

```python
raw_data
```

```python
data = np.array([cd for [cd,_y,f] in raw_data])
labels = np.array([_y for [cd,_y,f] in raw_data])
y = np.array([1 if label == 'men' else 0 for label in labels])
```

```python
data
```

```python
pca = RandomizedPCA(n_components=10, random_state=0)
X = pca.fit_transform(data)
```

```python
len(X[0])
```

```python

restored_image = pca.inverse_transform(X)
restored_image = np.clip(np.reshape(restored_image, (2484,70,70,3)), 0,255).astype(np.uint8)
restored_image = Image.fromarray(restored_image[0], 'RGB')
restored_image.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components = 1)
lda.fit(X,y)

```

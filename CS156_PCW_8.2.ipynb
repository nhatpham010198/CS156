```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style 
from sklearn import svm
from sklearn.model_selection import cross_val_score
```

```python
n = 1000
```

```python
dim1 = np.random.normal(0, 3, n)
dim2 = np.random.normal(5, 2, n)
```

```python
X = np.array([dim1, dim2]).T
X.shape
```

```python
y = [1 if np.random.random() < 0.5 else 0 for i in range(n)]
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

```python
clf = svm.SVC(kernel='linear', C = 1.0)
scores = cross_val_score(clf, X, y, cv = 5)
print(f"Score: {np.mean(scores)} +- {2*np.std(scores)}")
```

```python
for kernel in ('linear', 'poly', 'rbf'):
  clf = svm.SVC(kernel=kernel, C = 1.0)
  scores = cross_val_score(clf, X, y, cv = 5)
  print(f"Score for {kernel}: {round(np.mean(scores), 2)} +- {2*round(np.std(scores), 2)}")
```

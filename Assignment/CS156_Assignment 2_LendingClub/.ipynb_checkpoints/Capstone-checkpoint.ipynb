To improve the accuracy of the mode, I focus only on debt_consolidation loan (the major loan product of Lending Club) to avoid noises caused by different natures of various loan products.

```python
def merge_range(x,y):
    return('_'.join([str(x),str(y)]))
merge_range(5,6)
```

```python
accepted.dropna(inplace = True)
accepted['fico_range'] = accepted.apply(lambda x: merge_range(x.fico_range_low, x.fico_range_high), axis = 1)
accepted['fico_range']
```

```python
month_ordering = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
accepted['issue_m'] = accepted.issue_d.apply(lambda x: month_ordering.index(x.split('-')[0]))
accepted['issue_y'] = accepted.issue_d.apply(lambda x: x.split('-')[1])
```

```python
col_order = ['funded_amnt', 'loan_amnt', 'purpose', 'fico_range', 'dti', 'emp_length', 'issue_y', 'issue_m', 'addr_state']
accepted = accepted[col_order]
```

# Preprocess rejected data

```python
rejected = pd.read_csv('rejected_loan.csv')
rejected.dropna(inplace = True)
```

```python
#loan purpose in this data set is not well-written, so I need to use the following keyword to filter debt_conslidation loans
debt_consolidation = ['debt', 'Debt', 'consolidate', 'Consolidate', 'consolidating', 'Consolidating', 'consolidation', 'Consolidation']
def check_debt_consolidation(x, checklist = debt_consolidation):
    for i in checklist:
        if i in x:
            return True
    return(False)


#filter debt_consolidation loan from rejected data set
debt_conslidation_loan = [check_debt_consolidation(i) for i in rejected['Loan Title']]
rejected = rejected[debt_conslidation_loan]
```

```python
#we need to covert credit scores to range to allign with the structure of accepted data
def range_convert(x):
    lower = str((x//5)*5)
    upper = str((x//5)*5+4)
    return('_'.join([lower,upper]))


rejected['fico_range'] = rejected.Risk_Score.apply(lambda x: range_convert(x))
```

```python
#add funded_amount to align with the accepted data, and to use as the predicted variable
rejected['funded_amnt'] = 0
```

```python
#rename columns to align with the accepted data
rejected = rejected.rename(columns = {'Amount Requested':'loan_amnt', 'Loan Title':"purpose", 'Debt-To-Income Ratio':'dti', 'State':'addr_state', 'Employment Length':'emp_length'})
```

```python
#split application date column to year and month
rejected['issue_m'] = rejected['Application Date'].apply(lambda x: int(x.split('-')[1]))
rejected['issue_y'] = rejected['Application Date'].apply(lambda x: x.split('-')[0])
rejected.dti = rejected.dti.apply(lambda x: x[:-1])
```

```python
# rearrange column order to concatenate with accepted data
# col_order = ['funded_amnt', 'loan_amnt', 'purpose', 'fico_range', 'dti', 'emp_length', 'issue_y', 'issue_m', 'addr_state']
rejected = rejected[col_order]
```

```python
#as the rejected data is way larger than the accepted data, we just need a portion of it to avoid the skewness in output
rejected_sample = rejected.sample(n=accepted.shape[0], random_state = 111)
```

```python
# concatenate data
data = pd.concat([rejected_sample, accepted])
```

# Preprocess concatenated data

```python
#we train and test model on data_sub instead of data to avoid reloading the rejected dataset in case the code goes wrong
data_sub = data
```

```python
#convert emp_length to numerical values instead of using onehot_encode as longer working experience may imply better creditworthiness
emp_ordering = [
    '< 1 year',
    '1 year',
    '2 years',
    '3 years',
    '4 years',
    '5 years',
    '6 years',
    '7 years',
    '8 years',
    '9 years',
    '10+ years'
]

def ordinal_encode(df, column, ordering):
    df[column] = df[column].apply(lambda x: ordering.index(x))
    
ordinal_encode(data_sub, 'emp_length', emp_ordering)
```

```python
#use onehot_encode to cover state address as one state is not necessarily better than other in terms of creditworthiness
def onehot_encode(df, column):
    dummies = pd.get_dummies(df[column])
    df_new = pd.concat([df, dummies], axis=1)
    df_new = df_new.drop(column, axis=1)
    return df_new

data_sub = onehot_encode(data_sub, 'addr_state')
```

```python
#convert funded amount to range for the use of k-nearest neighbor below
def funded_range(x):
    lower = str((x//5000)*5000)
    upper = str((x//5000)*5000+4000)
    return('_'.join([lower,upper]))

data_sub['funded_range'] = data_sub['funded_amnt'].apply(lambda x: funded_range(x))
#the range of 40000-44000 also cover loans greater than 440000 as there was no debt consolidation loan that's worth more than 40000 in the company history
```

```python
#convert fico_range to number
data_sub['fico_num'] = data_sub['fico_range'].apply(lambda x: float(x.split('_')[0])//5)
```

```python
data_sub
```

```python
#drop unnecessary column
data_sub.drop(['funded_amnt', 'loan_amnt', 'fico_range', 'purpose'], axis = 1, inplace = True)
```

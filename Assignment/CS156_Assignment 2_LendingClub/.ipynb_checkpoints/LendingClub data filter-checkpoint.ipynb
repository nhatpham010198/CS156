From plots above, it is obvious that the company was growing rapidly from 2012 to 2016 and credit loan and debt consolidation loan are two main products of the company over time. For the best accuracy of the prediction, the following anlysis would focus only onf debt consolidation loan after 2012. 

The narrow focus on one loan product allows the model avoid noises caused by different natures of various loan products. 

The time span after 2012 is selected because 
* the 2008-2012 period was heavily subjected the effects of the financial crisis which may affect the mechanism of the company's current loan assessment model and 
* The model Lending Club used assess loan applicant changes over time. And the more recent data used to replicate that model the more accurate the replicated_model is.

```python
# var = pd.read_csv('final_var.csv')
# selected_var = var.Var
# print(selected_var.to_list())
selected_var = [ 'funded_amnt','term','emp_length', 'home_ownership', 'annual_inc', 'verification_status', 'addr_state', 'dti', 'delinq_2yrs', 'earliest_cr_line', 'fico_range_low', 'fico_range_high', 'inq_last_6mths', 'open_acc', 'pub_rec', 'revol_bal', 'revol_util', 'total_acc', 'initial_list_status', 'collections_12_mths_ex_med', 'application_type', 'tot_coll_amt', 'tot_cur_bal', 'open_acc_6m', 'open_act_il', 'open_il_12m', 'open_il_24m', 'total_bal_il', 'open_rv_12m', 'open_rv_24m', 'max_bal_bc', 'all_util', 'total_rev_hi_lim', 'inq_fi', 'total_cu_tl', 'inq_last_12m', 'acc_open_past_24mths', 'avg_cur_bal', 'bc_open_to_buy', 'bc_util', 'chargeoff_within_12_mths', 'delinq_amnt', 'mort_acc', 'num_accts_ever_120_pd', 'num_actv_bc_tl', 'num_actv_rev_tl', 'num_bc_sats', 'num_bc_tl', 'num_il_tl', 'num_op_rev_tl', 'num_rev_accts', 'num_rev_tl_bal_gt_0', 'num_sats', 'num_tl_120dpd_2m', 'num_tl_30dpd', 'num_tl_90g_dpd_24m', 'num_tl_op_past_12m', 'pct_tl_nvr_dlq', 'percent_bc_gt_75', 'pub_rec_bankruptcies', 'tax_liens', 'tot_hi_cred_lim', 'total_bal_ex_mort', 'total_bc_limit', 'total_il_high_credit_limit']
data = data[selected_var]
```

```python
data['earliest_cr_line_month'] = data.earliest_cr_line.apply(lambda x: x.split('-')[0])
data['earliest_cr_line_year'] = data.earliest_cr_line.apply(lambda x: x.split('-')[1])
data.drop('earliest_cr_line', axis = 1, inplace = True)
month_ordering = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
data['earliest_cr_line_month'] = data['earliest_cr_line_month'].apply(lambda x: month_ordering.index(x))
data
```

```python
for column in data.columns:
    try:
        data[column] = data[column].astype(np.float)
    except:
        pass
data
```

```python
cate_var_dict = {column: list(data[column].unique()) for column in data.columns if data.dtypes[column] == 'object'}
cate_var_dict
```

```python
cate_var = list(cate_var_dict.keys())
num_var = list(set(data.columns.values) - set(cate_var))
```

```python
data.emp_length[data.emp_length.isna()] = '1 year'
```

```python
binary_features = ['term','initial_list_status', 'application_type']
binary_positives = [' 60 months', 'w', 'Individual']

ordinal_features = ['emp_length']
emp_ordering = [
    '< 1 year',
    '1 year',
    '2 years',
    '3 years',
    '4 years',
    '5 years',
    '6 years',
    '7 years',
    '8 years',
    '9 years',
    '10+ years'
]

nominal_features = ['home_ownership', 'verification_status', 'addr_state']
```

```python
def binary_encode(df, column, positive_value):
    df[column] = df[column].apply(lambda x: 1 if x == positive_value else 0)

def ordinal_encode(df, column, ordering):
    df[column] = df[column].apply(lambda x: ordering.index(x))

def onehot_encode(df, column):
    dummies = pd.get_dummies(df[column])
    df_new = pd.concat([df, dummies], axis=1)
    df_new = df_new.drop(column, axis=1)
    return df_new
```

```python
for feature, positive_value in zip(binary_features, binary_positives):
    binary_encode(data, feature, positive_value)

ordinal_encode(data, 'emp_length', emp_ordering)

for feature in nominal_features:
    data = onehot_encode(data, feature)
```

```python
data.isna().mean().sort_values()
```

```python
data_nonna = data.dropna()
```

```python
data_nonna.isna().mean().sort_values()
```

```python
y = data_nonna.funded_amnt
X = data_nonna[num_var].drop('funded_amnt', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_predicted = reg.predict(X_test)
xaxis = range(0,y_test.shape[0])
plt.plot(xaxis[:200],y_test[:200])
plt.plot(xaxis[:200],y_predicted[:200])
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error
print('mean_squared_error is ', mean_squared_error(y_predicted,y_test))
print('mean_absolute_error is ', mean_absolute_error(y_predicted,y_test))
print('median_absolute_error is ',median_absolute_error(y_predicted,y_test))
```

```python
from sklearn.decomposition import PCA as RandomizedPCA
N_COMPONENTS = 50
pca = RandomizedPCA(n_components=N_COMPONENTS, random_state=0)
X1 = pca.fit_transform(X)
reg1 = LinearRegression()
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, train_size=0.7, random_state=42)
reg1.fit(X1_train,y_train)
predict1 = reg1.predict(X1_test)
print('mean_squared_error is ', mean_squared_error(y_test,predict1))
print('mean_absolute_error is ', mean_absolute_error(y_test,predict1))
print('median_absolute_error is ',median_absolute_error(y_test,predict1))
```
